```python
import pandas as pd

data = []

with open("/content/Genre Classification Dataset/train_data.txt", "r", encoding="utf-8") as file:
    for line in file:
        parts = line.strip().split(" ::: ")
        if len(parts) >= 4:
            data.append(parts[:4])

df = pd.DataFrame(data, columns=["ID", "Genre", "Title", "Plot"])

df.head()
```

```python
import matplotlib.pyplot as plt

genre_counts = df["Genre"].value_counts()

plt.figure(figsize=(12,6))
plt.bar(genre_counts.index, genre_counts.values)

plt.title("Genre Distribution")
plt.xlabel("Genre")
plt.ylabel("Number of Movies")
plt.xticks(rotation=90)

plt.tight_layout()
plt.show()
```
